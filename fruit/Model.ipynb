```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model
from IPython.display import Image
  
img_input=keras.Input(shape=(128,128,3),name="mango_input")
h1=layers.Dense(64,activation="relu",name="hidden1")(img_input)
h2=layers.Dense(64,activation="relu",name="hidden2")(h1)
output=layers.Dense(3,activation="softmax",name="output")(h2)
model=keras.Model(inputs=img_input,outputs=output)
plot_model(model,to_file="mango.png")
Image("mango.png")
```

```python
#取得所有圖片路徑
import os
dirPath = r"mango"
result = [f for f in os.listdir(dirPath) if os.path.isfile(os.path.join(dirPath, f))]
all_image_paths=["mango/"+f for f in result if f[-1]=="g"]    
image_count = len(all_image_paths)
print(all_image_paths[:5])
#顯示給你看看
import random
import IPython.display as display
import os
import pathlib


for n in range(3):
  image_path = random.choice(all_image_paths)
  display.display(display.Image(image_path))
  print()
#圖片解成tfrecord
def preprocess_image(image):
  image = tf.image.decode_jpeg(image, channels=3)
  image = tf.image.resize(image, [192, 192])
  image /= 255.0  # normalize to [0,1] range

  return image
def load_and_preprocess_image(path):
  image = tf.io.read_file(path)
  return preprocess_image(image)
import matplotlib.pyplot as plt

img_path = all_image_paths[0]

plt.imshow(load_and_preprocess_image(img_path))
plt.grid(False)
plt.xlabel(img_path)
print()
#image_ds是影像資料集，不含標籤的
path_ds = tf.data.Dataset.from_tensor_slices(all_image_paths)
AUTOTUNE = tf.data.experimental.AUTOTUNE
image_ds = path_ds.map(load_and_preprocess_image, num_parallel_calls=AUTOTUNE)
```
